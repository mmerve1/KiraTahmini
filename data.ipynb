```python
def drop_coloums(df,cols):
    ''''
    This function rumoves the coloums from a DataFrame
    '''
    for col in cols:
        try:
            df.drop([col],axis=1,inplace=True)
        except Exception as e:
            print(f"Bir hata olustu: {e}")
```
